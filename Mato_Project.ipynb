```python
import pandas as pd

# load all the data files
agency = pd.read_csv(r'C:\Users\PAWAN PATEL\Desktop\Python DS\Mrtro_analysis\DMRC_GTFS\agency.txt')
calendar = pd.read_csv(r'C:\Users\PAWAN PATEL\Desktop\Python DS\Mrtro_analysis\DMRC_GTFS\calendar.txt')                
routes = pd.read_csv(r'C:\Users\PAWAN PATEL\Desktop\Python DS\Mrtro_analysis\DMRC_GTFS\routes.txt')
shapes = pd.read_csv(r'C:\Users\PAWAN PATEL\Desktop\Python DS\Mrtro_analysis\DMRC_GTFS\shapes.txt')
stop_times = pd.read_csv(r'C:\Users\PAWAN PATEL\Desktop\Python DS\Mrtro_analysis\DMRC_GTFS\stop_times.txt')
stops = pd.read_csv(r'C:\Users\PAWAN PATEL\Desktop\Python DS\Mrtro_analysis\DMRC_GTFS\stops.txt')
trips = pd.read_csv(r'C:\Users\PAWAN PATEL\Desktop\Python DS\Mrtro_analysis\DMRC_GTFS\trips.txt')

# show the first few rows and the structure of each dataframe
data_overviews = {
    "agency": agency.head(),
    "calendar": calendar.head(),
    "routes": routes.head(),
    "shapes": shapes.head(),
    "stop_times": stop_times.head(),
    "stops": stops.head(),
    "trips": trips.head()
}
data_overviews
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.scatterplot(x='shape_pt_lon', y='shape_pt_lat', hue='shape_id', data=shapes, palette='viridis', s=10, legend=None)
plt.title('Geographical Paths of Delhi Metro Routes')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid(True)
plt.show()
```

```python
# merge trips with calendar to include the day of operation information
trips_calendar = pd.merge(trips, calendar, on='service_id', how='left')

# count the number of trips per day of the week
trip_counts = trips_calendar[['monday', 'tuesday', 'wednesday', 'thursday', 'friday', 'saturday', 'sunday']].sum()

# Plotting
plt.figure(figsize=(6, 5))
trip_counts.plot(kind='bar', color='skyblue')
plt.title('Number of Trips per Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Trips')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
